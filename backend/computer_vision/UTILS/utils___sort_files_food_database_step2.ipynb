```python
from PIL import Image
import os

# Function to convert PNG to RGBA and then fill transparency
def convert_image(file_path, output_path):
    with Image.open(file_path) as img:
        # Convert the image to RGBA
        img = img.convert('RGBA')

        # Create a new background image with white color and the same size
        background = Image.new('RGB', img.size, (255, 255, 255))
        
        # Paste the RGBA image onto the background (using alpha channel as mask)
        background.paste(img, mask=img.split()[3])  # 3 is the alpha channel

        # Save the image in the output path, you can choose format like 'JPEG' for non-transparency
        background.save(output_path, 'JPEG')

# Directories
base_dir = ".\\FOOD\\train"
test_base_dir = ".\\FOOD\\test"

# Loop through each class directory in both base and test directories
for _dir in [base_dir, test_base_dir]:
    for class_name in os.listdir(_dir):
        class_dir = os.path.join(_dir, class_name)
        if os.path.isdir(class_dir):
            filenames = os.listdir(class_dir)
            for f in filenames:
                if f.endswith('.png'):
                    file_path = os.path.join(class_dir, f)
                    print("Converting:", file_path)
                    
                    # Assuming you want to save the converted image in the same location
                    # with the original file name, but you can modify this as needed
                    output_path = os.path.join(class_dir, os.path.splitext(f)[0] + '.jpg')
                    
                    # Convert and save the image
                    convert_image(file_path, output_path)

print("Image conversion complete.")
```

```python
import os

def count_files_in_folders(directory):
    # Dictionary to store folder names and corresponding file counts
    folder_file_counts = {}

    # Loop through each folder in the specified directory
    for folder_name in os.listdir(directory):
        folder_path = os.path.join(directory, folder_name)
        # Check if the current item is a directory
        if os.path.isdir(folder_path):
            # Count the number of files in the current folder
            num_files = len([name for name in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, name))])
            # Store the folder name and file count in the dictionary
            folder_file_counts[folder_name] = num_files

    return folder_file_counts

def rename_folders(directory):
    # Get folder counts
    folder_file_counts = count_files_in_folders(directory)

    # Rename folders with less than 100 files
    for folder, count in folder_file_counts.items():
        if count < 150:
            new_folder_name = f"_{count}_{folder}"
            old_folder_path = os.path.join(directory, folder)
            new_folder_path = os.path.join(directory, new_folder_name)
            os.rename(old_folder_path, new_folder_path)
            print(f"Renamed folder '{folder}' with {count} files to '{new_folder_name}'.")

# Example usage:
directory_path = "C:\\Users\\ijzep\\_IA_LAMBTON\\Capstone\\_part2_image_classification\\FOOD\\train"  # Specify the directory path here
rename_folders(directory_path)
```

```python
import os
import re

def rename_folders_with_format(directory):
    # Loop through each folder in the specified directory
    for folder_name in os.listdir(directory):
        folder_path = os.path.join(directory, folder_name)
        # Check if the current item is a directory
        if os.path.isdir(folder_path):
            # Check if the folder name matches the specified format
            match = re.match(r'^_(\d+)_(.*)$', folder_name)
            if match:
                # Extract the name part from the folder name
                new_folder_name = match.group(2)
                # Construct the new folder path
                new_folder_path = os.path.join(directory, new_folder_name)
                # Rename the folder
                os.rename(folder_path, new_folder_path)
                print(f"Renamed folder '{folder_name}' to '{new_folder_name}'.")
            else:
                print(f"Folder '{folder_name}' does not match the specified format and will be left unchanged.")

# Example usage:
directory_path = "C:\\Users\\ijzep\\_IA_LAMBTON\\Capstone\\_part2_image_classification\\FOOD\\train"  # Specify the directory path here
rename_folders_with_format(directory_path)
```

```python
import os
import random
import shutil

def split_train_test_data(dataset_dir, train_ratio=0.8):
    # Create train and test directories at the root level of the dataset directory
    train_dir = os.path.join(dataset_dir, "train")
    test_dir = os.path.join(dataset_dir, "test")
    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(test_dir, exist_ok=True)

    # Loop through each class folder in the dataset directory
    for class_name in os.listdir(dataset_dir):
        class_dir = os.path.join(dataset_dir, class_name)
        # Make sure it's a directory
        if os.path.isdir(class_dir):
            # Remove the train and test folders if they exist inside the class folder
            shutil.rmtree(os.path.join(class_dir, "train"), ignore_errors=True)
            shutil.rmtree(os.path.join(class_dir, "test"), ignore_errors=True)

            # Get list of image filenames in the class folder
            image_files = [f for f in os.listdir(class_dir) if f.endswith(('.jpg', '.jpeg', '.png', '.gif'))]
            # Shuffle the list of filenames randomly
            random.shuffle(image_files)
            
            # Determine the number of images for training and testing
            num_train = int(len(image_files) * train_ratio)
            num_test = len(image_files) - num_train

            # Create train and test directories within the class folder
            class_train_dir = os.path.join(train_dir, class_name)
            class_test_dir = os.path.join(test_dir, class_name)
            os.makedirs(class_train_dir, exist_ok=True)
            os.makedirs(class_test_dir, exist_ok=True)

            # Copy images to train directory
            for img_file in image_files[:num_train]:
                src = os.path.join(class_dir, img_file)
                dst = os.path.join(class_train_dir, img_file)
                shutil.copy(src, dst)
                
            # Copy images to test directory
            for img_file in image_files[num_train:]:
                src = os.path.join(class_dir, img_file)
                dst = os.path.join(class_test_dir, img_file)
                shutil.copy(src, dst)
                
            print(f"Split data for class '{class_name}': {num_train} images for training, {num_test} images for testing.")


dataset_directory = "C:\\Users\\ijzep\\_IA_LAMBTON\\Capstone\\_part2_image_classification\\FOOD\\full"  # Specify the directory path to your dataset here
split_train_test_data(dataset_directory)
```

```python
## Checar las extensiones
import os
from collections import defaultdict

def list_file_extensions(directory):
    """
    List all file extensions in the given directory and its subdirectories.

    Args:
    - directory: The path to the directory to scan.

    Returns:
    A dictionary where keys are file extensions and values are the counts of files with that extension.
    """
    extensions_count = defaultdict(int)
    for root, dirs, files in os.walk(directory):
        for file in files:
            extension = os.path.splitext(file)[1].lower()
            extensions_count[extension] += 1
    return extensions_count

directory_path = "C:\\Users\\ijzep\\_IA_LAMBTON\\Capstone\\_part2_image_classification\\FOOD\\test"
extensions = list_file_extensions(directory_path)
if extensions:
    print("File extensions found and their counts:")
    for ext, count in extensions.items():
        print(f"{ext}: {count}")
else:
    print("No files found in the specified directory.")
```

```python
# parte 2
import os
from PIL import Image
import warnings

def find_corrupted_images(directory):
    """
    Encuentra y lista las imágenes que no pueden ser abiertas o generan advertencias al intentar ser abiertas en el directorio especificado.
    """
    corrupted_files = []
    warnings.filterwarnings('error', category=UserWarning, message='Truncated File Read')  # Convierte la advertencia en excepción

    for root, dirs, files in os.walk(directory):
        for file in files:
            if str(file).lower().endswith(('.png', '.jpg', '.jpeg', '.tiff', '.tif')):  # Agregado soporte para archivos TIFF
                try:
                    img_path = os.path.join(root, file)
                    with Image.open(img_path) as img:
                        img.verify()  # Verifica la imagen
                except (IOError, SyntaxError, UserWarning) as e:  # Captura la excepción de advertencia
                    print(f"Archivo potencialmente dañado o incompleto: {img_path}")
                    corrupted_files.append(img_path)

    warnings.resetwarnings()  # Restablece el comportamiento predeterminado de las advertencias
    return corrupted_files
 
test_directory = "C:\\Users\\ijzep\\_IA_LAMBTON\\Capstone\\_part2_image_classification\\FOOD\\test"   # Ajusta esta ruta al directorio correcto
damaged_files = find_corrupted_images(test_directory)
if damaged_files:
    print(f"Se encontraron {len(damaged_files)} archivos potencialmente dañados o incompletos.")
else:
    print("No se encontraron archivos dañados o incompletos.")
```
